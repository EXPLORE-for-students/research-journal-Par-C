The high resolution simulation was compared to the low resolution simulation, and they both do not fit to the following equations (2):
$$
\frac{M_\bullet}{M_\star} \propto \xi(z)^{5/6} (1 + z)^{5/2}.
$$

$$
\mathcal{E}(z) = \frac{\xi(z)^{5/6}(1 + z)^{5/2}}{\xi(0)^{5/6}}.
$$

This consistency of the data, and the consistency of these equations not fitting suggests that these supermassive blackholes at high redshift do not conform to the expectations of low red shift blackholes.
